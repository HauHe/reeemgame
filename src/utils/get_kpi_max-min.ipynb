```python
import json
import pandas as pd
```

```python
kpis = ['AccumulatedCO2','CO2Intensity', 'DiscountedInvestmentPerCitizen', 'LCOE']
kpi_dic = {}
for kpi in kpis:
    path = '../data/indicators/' + kpi + '.json'
    kpi_dic[kpi] = pd.read_json(path)
```

```python
kpi_dic['AccumulatedCO2'].head()
```

```python
countries = kpi_dic['LCOE'].iloc[0, 0].keys()
```

```python
kpi_df_dic = {}
for kpi in kpis:
    kpi_df_dic[kpi] = pd.DataFrame(columns=['scenario', 'year','region','value'])
    for c in kpi_dic[kpi].columns:
        df_0 = pd.DataFrame(columns=['scenario', 'year','region','value'])
        for r in kpi_dic[kpi].index:
            df_0['region'] = pd.Series(countries)
            df_0['value'] = pd.Series(map(kpi_dic[kpi].loc[r, c].get, countries))
            df_0['scenario'] = c
            df_0['year'] = r
            kpi_df_dic[kpi] = pd.concat([kpi_df_dic[kpi], df_0], ignore_index=True)
```

```python
df_kpi_stats = pd.DataFrame(columns=['kpi', 'stat', 'value'])
for kpi in kpis:
    max = kpi_df_dic[kpi]['value'].max()
    min = kpi_df_dic[kpi]['value'].min()
    df_kpi_stats = pd.concat([df_kpi_stats, pd.DataFrame([[kpi, 'max', max]], columns=['kpi', 'stat', 'value'])], ignore_index=True)
    df_kpi_stats = pd.concat([df_kpi_stats, pd.DataFrame([[kpi, 'min', min]], columns=['kpi', 'stat', 'value'])], ignore_index=True)
print(df_kpi_stats)
```
